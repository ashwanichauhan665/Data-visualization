```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cricket_analysis_dataset.csv')
df = pd.DataFrame(data)
df.head()
```

```python
pd.isnull(df)
```

```python
df = df.fillna(value=0)
```

```python
Runs = df["Runs Scored"].astype(int)
Economy_Rate = df["Economy Rate"].astype(int)
Wicket_taken = df["Wickets Taken"].astype(int)
Batting_average = df["Batting Average"].astype(int)
Bowling_average = df["Bowling Average"].astype(int)
Strike_Rate = df["Strike Rate"].astype(float)
```

<h1>Batting Analysis</h1>


```python
player_data = df.groupby(['Player Name'])['Runs Scored'].sum().head(50)

plt.figure(figsize=(10,10))
plt.barh(player_data.index,player_data.values, color='purple',height=0.4)
plt.xlabel('Runs scored')
plt.ylabel('Player name')
plt.title('Runs Scored by Each Player')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_10_scorers = df.sort_values(by='Runs Scored', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_scorers['Player Name'], top_10_scorers['Runs Scored'], color='green',height=0.4)
plt.xlabel('Runs Scored')
plt.ylabel('Player Name')
plt.title('Top 10 Run Scorers')
plt.gca().invert_yaxis() 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Strike Rate'], df['Batting Average'], color='purple')
plt.xlabel('Strike Rate')
plt.ylabel('Batting Average')
plt.title('Batting Average vs Strike Rate')
plt.grid(True)
plt.show()
```

```python
player_Names = df.groupby(['Player Name'])['Batting Average'].sum().head(15)

plt.figure(figsize=(10, 6))
plt.plot(player_Names.index,player_Names.values, marker='o', linestyle='-', color='b')
plt.xlabel('Players')
plt.ylabel('Batting Average')
plt.title('Batting Average of Each Player')
plt.xticks(rotation=45)
plt.show()
```

```python
top_20_df = df.head(20)

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))
bar_width = 0.35
index = np.arange(len(top_20_df))

bars1 = ax.bar(index - bar_width/2, top_20_df['Centuries'], bar_width, label='Centuries', color='blue')
bars2 = ax.bar(index + bar_width/2, top_20_df['Half Centuries'], bar_width, label='Half Centuries', color='green')

ax.set_xlabel('Players')
ax.set_ylabel('Counts')
ax.set_title('Centuries and Half Centuries')
ax.set_xticks(index)
ax.set_xticklabels(top_20_df['Player Name'], rotation=90, fontsize=8)
ax.legend()

plt.tight_layout()
plt.show()
```

<h1>Bowling Analysis</h1>

```python
wickets_by_player = df.groupby('Player Name')['Wickets Taken'].sum().sort_values(ascending=False).head(10)
def func(pct, allvalues):
    absolute = int(pct/100.*sum(allvalues))
    return f"{absolute} wickets\n({pct:.1f}%)"

plt.figure(figsize=(8, 8))
plt.pie(wickets_by_player, labels=wickets_by_player.index, autopct=lambda pct: func(pct, wickets_by_player), startangle=140)
plt.title('Wickets Taken by Top 10 Players')
plt.show()
```

```python
top_20_df = df.head(20)

# Plotting Runs Scored
fig, ax = plt.subplots(figsize=(10, 8))
bar_width = 0.5
index = np.arange(len(top_20_df))

bars = ax.bar(index, top_20_df['Runs Scored'], bar_width, color='blue', label='Runs Scored')

ax.set_xlabel('Players')
ax.set_ylabel('Runs Scored')
ax.set_title('Top 20 Players by Runs Scored')
ax.set_xticks(index)
ax.set_xticklabels(top_20_df['Player Name'], rotation=90, fontsize=8)
ax.legend()

plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.bar(index, top_20_df['Wickets Taken'], bar_width, color='green', label='Wickets Taken')

ax.set_xlabel('Players')
ax.set_ylabel('Wickets Taken')
ax.set_title('Top 20 Players by Wickets Taken')
ax.set_xticks(index)
ax.set_xticklabels(top_20_df['Player Name'], rotation=90, fontsize=8)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
wickets_by_players = df.groupby('Player Name')['Wickets Taken'].sum().head(40)

plt.figure(figsize=(10, 8))
wickets_by_players.sort_values().plot(kind='barh', color='salmon')
plt.xlabel('Wickets Taken')
plt.ylabel('Player Name')
plt.title('Wickets Taken by Each Player')
plt.show()
```

```python
filtered_df = df[df['5 Wicket Hauls'] > 0]
agg_df = filtered_df.groupby('5 Wicket Hauls').apply(lambda x: ', '.join(x['Player Name'])).reset_index(name='Players')
sizes = filtered_df['5 Wicket Hauls'].value_counts().sort_index()
labels = [f'{agg_df.loc[agg_df["5 Wicket Hauls"] == label, "Players"].values[0]}' for label, count in sizes.items()]
colors = plt.get_cmap('tab20').colors[:len(sizes)] 
def func(pct, allvals):
    absolute = int(pct / 100. * sum(allvals))
    return f'{absolute}'

fig, ax = plt.subplots(figsize=(8, 8))  
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct=lambda pct: func(pct, sizes), startangle=140)
ax.set_title('Distribution of 5 Wicket Hauls ', fontsize=14)

for text in texts:
    text.set_fontsize(10)  
for autotext in autotexts:
    autotext.set_fontsize(12)  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Runs Scored'], df['Wickets Taken'], s=df['Strike Rate'], alpha=0.5, c='red', edgecolors='w', linewidth=2)
plt.xlabel('Runs Scored')
plt.ylabel('Wickets Taken')
plt.title('Bubble Chart of Runs Scored vs Wickets Taken')
plt.show()
```

<h1>Feilding Analysis</h1>

```python
fig, ax = plt.subplots(figsize=(14, 6))

# Plotting Catches
ax.barh(top_20_df['Player Name'], top_20_df['Cathes'], label='Cathes', color='blue', height=0.4)

# Plotting Run Outs
ax.barh(top_20_df['Player Name'], top_20_df['Run Out'], label='Run Out', color='red', left=top_20_df['Cathes'], height=0.4)

# Labels and Titles
ax.set_xlabel('Counts')
ax.set_ylabel('Players')
ax.set_title('Catches and Run Outs')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
df['Total Fielding Contribution'] = df['Cathes'] + df['Run Out']

top_10_fielders = df.sort_values(by='Total Fielding Contribution', ascending=False).head(10)
bar_width = 0.35
index = np.arange(len(top_10_fielders))

plt.figure(figsize=(12, 8))
plt.bar(index, top_10_fielders['Cathes'], bar_width, label='Cathes', color='skyblue')
plt.bar(index + bar_width, top_10_fielders['Run Out'], bar_width, label='Run Outs', color='orange')

plt.xlabel('Player Name')
plt.ylabel('Fielding Contribution')
plt.title('Top 10 Fielders Based on Catches and Run Outs')
plt.xticks(index + bar_width / 2, top_10_fielders['Player Name'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
