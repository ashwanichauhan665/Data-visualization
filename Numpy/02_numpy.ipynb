```python
import numpy as np

# Generate random temperature data for a week (7 days * 24 hours)
np.random.seed(0)  # for reproducibility
temperature_data = np.random.uniform(low=-10, high=35, size=(7, 24))
print("Temperature data (first day):")
print(temperature_data[0])

# Get temperature data for the first day
first_day = temperature_data[0]
print("\nTemperature data for the first day:")
print(first_day)

# Get temperature at noon (12 PM) for all days
noon_temps = temperature_data[:, 12]
print("\nTemperature at noon for all days:")
print(noon_temps)

# Get temperature data for the first three days
first_three_days = temperature_data[:3]
print("\nTemperature data for the first three days:")
print(first_three_days)

# Reshape the data to have 2 weeks of data with 12-hour intervals
reshaped_data = temperature_data.reshape((2, 7, 12))
print("\nReshaped data (2 weeks of 12-hour intervals):")
print(reshaped_data)

# Flatten the data to a 1D array
flattened_data = temperature_data.flatten()
print("\nFlattened temperature data:")
print(flattened_data)

# Calculate the mean temperature for the week
mean_temp = np.mean(temperature_data)
print("\nMean temperature for the week:", mean_temp)

# Calculate the maximum temperature for the week
max_temp = np.max(temperature_data)
print("Maximum temperature for the week:", max_temp)

# Calculate the minimum temperature for the week
min_temp = np.min(temperature_data)
```
